## 2 Illustrative Example: Predicting Risk of Ischemic Stroke

```python
import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from scipy.stats import chi2_contingency
from scipy.stats import ttest_ind

from sklearn.preprocessing import PowerTransformer
from sklearn.preprocessing import StandardScaler

from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV

from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline
```

```python
train = pd.read_csv("stroke_train.csv", index_col=0)
test = pd.read_csv("stroke_test.csv", index_col=0)

data = pd.concat([train, test]).sort_index()
data.head()
```

```python
blockage = data["NASCET"].map({0: "Blockage < 70%", 1: "Blockage > 70%"})
blockage = blockage.rename("Blockage")
stroke = data["Stroke"].map({"N": "Stroke=No", "Y": "Stroke=Yes"})

contingency = pd.crosstab(blockage, stroke)
contingency
```

```python
chi2, p, dof, expected = chi2_contingency(contingency)
print("Chi-square stat: ", chi2)
print("P-value: ", p)
```

## 2.2 Preprocessing

```python
VC_preds = [
    "MaxLRNCAreaProp", "MaxLRNCArea", "LRNCVol", "LRNCVolProp",
    "MATXVol", "WallVol", "MaxMATXArea", "MaxWallArea",
    "MaxDilationByArea", "MaxRemodelingRatio", "MATXVolProp", "MaxMATXAreaProp",
    "MaxCALCAreaProp", "MaxCALCArea", "CALCVol", "CALCVolProp",
    "MaxMaxWallThickness", "MaxStenosisByArea", "MaxStenosisByDiameter"]

risk_preds = [
    "age", "sex", "SmokingHistory", "AtrialFibrillation", "CoronaryArteryDisease", 
    "DiabetesHistory", "HypercholesterolemiaHistory", "HypertensionHistory"]

response_var = "Stroke"
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 7))
# Initial MaxLRNCArea distribution
sns.histplot(data["MaxLRNCArea"], ax=ax1, bins=15, color="#D53E4F")

ss = StandardScaler()
yj_std = PowerTransformer(method='yeo-johnson', standardize=True)

transformed_MaxLRNCArea = yj_std.fit_transform(data["MaxLRNCArea"].values.reshape((-1,1)))
ax2.set_xlabel("Transformed MaxLRNCArea")
sns.histplot(transformed_MaxLRNCArea.ravel(), ax=ax2, bins=15, color="#D53E4F");
```

```python
transformed_VC_preds = yj_std.fit_transform(train[VC_preds])
transformed_VC_preds_df = pd.DataFrame(transformed_VC_preds, columns=VC_preds)
corr = transformed_VC_preds_df.corr()
cmap = sns.diverging_palette(15, 250, s=100, l=30, as_cmap=True)
f, ax = plt.subplots(figsize=(15, 9))
sns.heatmap(corr, cmap=cmap, center=0,
            square=True, linewidths=.5, vmin=-1, vmax=1, cbar_kws={"shrink": .5});
```

```python
corr_matrix = corr.abs()
corr_threshold = 0.9 # <-- arbitrary
high_corr_var = np.where(corr_matrix > corr_threshold)

# Columns with high correlation
[(corr_matrix.columns[x],corr_matrix.columns[y]) for x,y in zip(*high_corr_var) if x!=y and x<y]
```

## 2.3 Exploration

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=5)
dc = DummyClassifier(strategy='stratified')
lr = LogisticRegression(max_iter=400, solver='newton-cg')

train[response_var] = train[response_var].map({'N': 0, 'Y': 1})
test[response_var] = test[response_var].map({'N': 0, 'Y': 1})
```

```python
risk_scores = []
for pred in risk_preds:
    dc_cv_scores = cross_val_score(dc, train[[pred]], train[response_var], scoring="roc_auc", cv=cv)
    lr_cv_scores = cross_val_score(lr, train[[pred]], train[response_var], scoring="roc_auc", cv=cv)

    t_stat, p_value = ttest_ind(lr_cv_scores, dc_cv_scores)
    diff = lr_cv_scores.mean() - dc_cv_scores.mean()
    one_sided_p = p_value/2 if diff > 0 else 1 - p_value/2
    
    risk_scores.append({
        "Predictor": pred,
        "Improvement": diff,
        "Pvalue": one_sided_p,
        "ROC": lr_cv_scores.mean()
    })
```

```python
risk_scores_df = pd.DataFrame(risk_scores)
risk_scores_df.sort_values(by="Pvalue", ascending=True)
```

```python
pipeline_log_reg = Pipeline(steps=[("yj-transform", yj_std), ("logistic-regression", lr)])
vc_scores = []
for pred in VC_preds:
    dc_cv_scores = cross_val_score(dc, train[[pred]], train[response_var], scoring="roc_auc", cv=cv)
    lr_cv_scores = cross_val_score(pipeline_log_reg, train[[pred]], train[response_var], scoring="roc_auc", cv=cv)

    t_stat, p_value = ttest_ind(lr_cv_scores, dc_cv_scores)
    diff = lr_cv_scores.mean() - dc_cv_scores.mean()
    one_sided_p = p_value/2 if diff > 0 else 1 - p_value/2
    
    vc_scores.append({
        "Predictor": pred,
        "Pvalue": one_sided_p,
        "ROC": lr_cv_scores.mean()
    })
```

```python
vc_scores_df = pd.DataFrame(vc_scores).sort_values(by="Pvalue")
vc_scores_df
```

```python
transformed_VC_preds_df = pd.DataFrame(yj_std.fit_transform(train[VC_preds]), columns=VC_preds)
transformed_VC_preds_df["Stroke"] = train["Stroke"]
fig, axes = plt.subplots(round(len(VC_preds)/4), 4, figsize=(15, 20))
for idx, pred in enumerate(vc_scores_df["Predictor"]):
    i, j = divmod(idx, 4)
    ax = axes[i, j]
    p_val = vc_scores_df[vc_scores_df["Predictor"] == pred]["Pvalue"]
    ax.set_title(pred + ", P-value: %.5f" % p_val)
    ax.set_ylabel(" ")
    sns.boxplot(x=response_var, y=pred, data=transformed_VC_preds_df, ax=ax)
    
fig.tight_layout();
```

```python
lr.fit(train[["MaxRemodelingRatio"]], train[response_var])
fpr, tpr, thresholds = roc_curve(train[response_var], lr.predict_proba(train[["MaxRemodelingRatio"]])[:, 1])
roc_auc = auc(fpr, tpr)

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
train_scaled = train.copy()
train_scaled[VC_preds] = yj_std.fit_transform(train[VC_preds])

VC_pairs = itertools.combinations(VC_preds, 2)
vc_int_scores = []
for pred1, pred2 in VC_pairs:
    # Model with just a pair of vascuCAP predictors
    lr_cv_scores = cross_val_score(lr, train_scaled[[pred1, pred2]], train_scaled[response_var], scoring="roc_auc", cv=cv)
    
    # Model with a pair of vascuCAP predictors + an interaction term
    train_scaled[pred1 + "_" + pred2] = train_scaled[pred1] * train_scaled[pred2]
    lr_int_cv_scores = cross_val_score(lr, train_scaled[[pred1, pred2, pred1 + "_" + pred2]], train_scaled[response_var], scoring="roc_auc", cv=cv)
    
    t_stat, p_value = ttest_ind(lr_int_cv_scores, lr_cv_scores)
    diff = lr_int_cv_scores.mean() - lr_cv_scores.mean()
    one_sided_p = p_value/2 if diff > 0 else 1 - p_value/2
    
    vc_int_scores.append({
        "Predictor1": pred1,
        "Predictor2": pred2,
        "Improvement": diff,
        "Pvalue": one_sided_p,
        "ROC_base": lr_cv_scores.mean(),
        "ROC_int": lr_int_cv_scores.mean()
    })
```

```python
vc_int_scores_df = pd.DataFrame(vc_int_scores).sort_values(by="Pvalue")
```

## 2.4 Predictive Modeling Across Sets

```python
def add_get_interactions(df, preds_for_int, pairs = []):
    int_preds = []
    if len(pairs) == 0:
        pairs = list(itertools.combinations(preds_for_int, 2))
    for pred1, pred2 in pairs:
        new_int_pred = pred1 + "_" + pred2
        int_preds.append(new_int_pred)
        df[new_int_pred] = df[pred1] * df[pred2]
            
    return int_preds
```

```python
def get_uncorrelated_columns(df, threshold = 0.75):
    df_corr = df.corr()
    columns = np.full((df_corr.shape[0],), True, dtype=bool)
    for i in range(df_corr.shape[0]):
        for j in range(i+1, df_corr.shape[0]):
            if df_corr.iloc[i,j] >= threshold:
                if columns[j]:
                    columns[j] = False
    uncorr_columns = df.columns[columns]
    return uncorr_columns
```

```python
def rfe_eval(df, preds, response_var):            
    result = {
        "Features": [],
        "ROC": []
    }
    for i in range(1, len(preds)+1):
        rfe = RFE(estimator=lr, n_features_to_select=i)
        scores = cross_val_score(rfe, df[risk_preds], df[response_var], scoring="roc_auc", cv=cv)
        result["Features"].append(i)
        result["ROC"].append(scores.mean())
        
    return pd.DataFrame(result)
```

```python
# Risk preds, main effects, no corr filter
train_copy = train.copy()
train_copy[risk_preds] = ss.fit_transform(train_copy[risk_preds])
risk_main = rfe_eval(train_copy, risk_preds, response_var)
```

```python
# Risk preds, main effects + int, no corr filter
train_copy = train.copy()
int_preds = add_get_interactions(train_copy, risk_preds)
train_copy[risk_preds + int_preds] = ss.fit_transform(train_copy[risk_preds + int_preds])
risk_int = rfe_eval(train_copy, risk_preds + int_preds, response_var)
```

```python
# Risk preds, main effects, w/ corr filter
train_copy = train.copy()
uncorr_preds = get_uncorrelated_columns(train_copy[risk_preds])
train_copy[uncorr_preds] = ss.fit_transform(train_copy[uncorr_preds])
risk_main_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
# Risk preds, main effects + int, w/ corr filter
train_copy = train.copy()
int_preds = add_get_interactions(train_copy, risk_preds)
uncorr_preds = get_uncorrelated_columns(train_copy[risk_preds + int_preds])
train_copy[uncorr_preds] = ss.fit_transform(train_copy[uncorr_preds])
risk_int_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
# VC preds, main effects, no corr filter
train_copy = train.copy()
train_copy[VC_preds] = yj_std.fit_transform(train_copy[VC_preds])
vc_main = rfe_eval(train_copy, VC_preds, response_var)
```

```python
# VC preds, main effects + int, no corr filter
train_copy = train.copy()
filtered_vc_ints = vc_int_scores_df[(vc_int_scores_df["Pvalue"] <= 0.2)]
pairs = list(zip(filtered_vc_ints["Predictor1"], filtered_vc_ints["Predictor2"]))
int_preds = add_get_interactions(train_copy, [], pairs)
# train_copy[VC_preds + int_preds] = ss.fit_transform(train_copy[VC_preds + int_preds])
train_copy[VC_preds + int_preds] = yj_std.fit_transform(train_copy[VC_preds + int_preds])
vc_int = rfe_eval(train_copy, VC_preds + int_preds, response_var)
```

```python
# VC preds, main effects, w/ corr filter
train_copy = train.copy()
uncorr_preds = get_uncorrelated_columns(train_copy[VC_preds])
train_copy[uncorr_preds] = yj_std.fit_transform(train_copy[uncorr_preds])
vc_main_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
# VC preds, main effects + int, w/ corr filter
train_copy = train.copy()
filtered_vc_ints = vc_int_scores_df[(vc_int_scores_df["Pvalue"] <= 0.2)]
pairs = list(zip(filtered_vc_ints["Predictor1"], filtered_vc_ints["Predictor2"]))
int_preds = add_get_interactions(train_copy, [], pairs)
uncorr_preds = get_uncorrelated_columns(train_copy[VC_preds + int_preds])
# train_copy[uncorr_preds] = ss.fit_transform(train_copy[uncorr_preds])
train_copy[uncorr_preds] = yj_std.fit_transform(train_copy[uncorr_preds])
vc_int_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
# All preds, main effects, no corr filter
train_copy = train.copy()
# train_copy[risk_preds] = ss.fit_transform(train_copy[risk_preds])
train_copy[VC_preds] = yj_std.fit_transform(train_copy[VC_preds])
all_main = rfe_eval(train_copy, risk_preds + VC_preds, response_var)
```

```python
# All preds, main effects + int, no corr filter
train_copy = train.copy()
filtered_vc_ints = vc_int_scores_df[(vc_int_scores_df["Pvalue"] <= 0.2)]
pairs = list(zip(filtered_vc_ints["Predictor1"], filtered_vc_ints["Predictor2"]))
int_preds = add_get_interactions(train_copy, [], pairs)
# train_copy[risk_preds + VC_preds + int_preds] = ss.fit_transform(train_copy[risk_preds + VC_preds + int_preds])
train_copy[risk_preds] = ss.fit_transform(train_copy[risk_preds])
train_copy[VC_preds + int_preds] = yj_std.fit_transform(train_copy[VC_preds + int_preds])
all_int = rfe_eval(train_copy, risk_preds + VC_preds + int_preds, response_var)
```

```python
# All preds, main effects, w/ corr filter
train_copy = train.copy()
uncorr_preds = get_uncorrelated_columns(train_copy[risk_preds + VC_preds])
# train_copy[uncorr_preds] = ss.fit_transform(train_copy[uncorr_preds])
risk_uncorr_preds = [x for x in uncorr_preds if x in risk_preds]
train_copy[risk_uncorr_preds] = ss.fit_transform(train_copy[risk_uncorr_preds])
VC_uncorr_preds = [x for x in uncorr_preds if x in VC_preds]
train_copy[VC_uncorr_preds] = yj_std.fit_transform(train_copy[VC_uncorr_preds])
all_main_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
# All preds, main effects + int, w/ corr filter
train_copy = train.copy()
filtered_vc_ints = vc_int_scores_df[(vc_int_scores_df["Pvalue"] <= 0.2)]
pairs = list(zip(filtered_vc_ints["Predictor1"], filtered_vc_ints["Predictor2"]))
int_preds = add_get_interactions(train_copy, [], pairs)
uncorr_preds = get_uncorrelated_columns(train_copy[risk_preds + VC_preds + int_preds])
# train_copy[uncorr_preds] = ss.fit_transform(train_copy[uncorr_preds])
risk_uncorr_preds = [x for x in uncorr_preds if x in risk_preds]
train_copy[risk_uncorr_preds] = ss.fit_transform(train_copy[risk_uncorr_preds])
VC_uncorr_preds = [x for x in uncorr_preds if (x in VC_preds) or (x in int_preds)]
train_copy[VC_uncorr_preds] = yj_std.fit_transform(train_copy[VC_uncorr_preds])
all_int_filter = rfe_eval(train_copy, uncorr_preds, response_var)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15,8))
# Risk preds
sns.lineplot(x="Features", y="ROC", data=risk_main, ax=axes[0,0], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=risk_int, ax=axes[0,0], palette=("blue",))

sns.lineplot(x="Features", y="ROC", data=risk_main_filter, ax=axes[1,0], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=risk_int_filter, ax=axes[1,0], palette=("blue",))

# VC preds
sns.lineplot(x="Features", y="ROC", data=vc_main, ax=axes[0,1], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=vc_int, ax=axes[0,1], palette=("blue",))

sns.lineplot(x="Features", y="ROC", data=vc_main_filter, ax=axes[1,1], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=vc_int_filter, ax=axes[1,1], palette=("blue",))

# All preds
sns.lineplot(x="Features", y="ROC", data=all_main, ax=axes[0,2], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=all_int, ax=axes[0,2], palette=("blue",))

sns.lineplot(x="Features", y="ROC", data=all_main_filter, ax=axes[1,2], palette=("red",))
sns.lineplot(x="Features", y="ROC", data=all_int_filter, ax=axes[1,2], palette=("blue",))


axes[0,0].set_ylim([0.4, 0.7])
axes[1,0].set_ylim([0.4, 0.7])
axes[0,1].set_ylim([0.4, 0.7])
axes[1,1].set_ylim([0.4, 0.7])
axes[0,2].set_ylim([0.4, 0.7])
axes[1,2].set_ylim([0.4, 0.7])

axes[0,0].set_xlim([0, 55])
axes[1,0].set_xlim([0, 55])
axes[0,1].set_xlim([0, 55])
axes[1,1].set_xlim([0, 55])
axes[0,2].set_xlim([0, 55])
axes[1,2].set_xlim([0, 55])

axes[0,0].set_title("Risk Predictors | No filter")
axes[1,0].set_title("Risk Predictors | With filter")
axes[0,1].set_title("VC Predictors | No filter")
axes[1,1].set_title("VC Predictors | With filter")
axes[0,2].set_title("All Predictors | No filter")
axes[1,2].set_title("All Predictors | With filter")

fig.tight_layout();
```
